```python
# Dropping unnecessary columns
X.drop(columns=["RowNumber","CustomerId","Surname"], inplace =True)
```

```python
X.head()
```

```python
cat_cols = X.select_dtypes(["object"])
for i in cat_cols.columns:
    dummy = pd.get_dummies(X[i], drop_first=True)
    X = pd.concat([X,dummy], axis=1)
    X = X.drop(columns=i)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print("trainning and test independent variables size:" f"{X_train.shape, X_test.shape}")
print("trainning and test dependent variables size:" f"{y_train.shape, y_test.shape}")
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
from tensorflow import keras
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
from keras.models import Sequential # for creating  
from keras.layers import Dense, Activation, Embedding, Flatten, LeakyReLU, BatchNormalization, Dropout # importing layers
from keras.activations import relu, sigmoid # importing activation functions 
```

```python
def create_model(layers, activation): # two variables layer and activations
    model = Sequential() # creating the model pipeline
    for i, nodes in enumerate(layers):
        if i==0: # for first layer we have to input the number of features
            model.add(Dense(nodes, input_dim=X_train.shape[1])) # nodes = iterating, input_dim = features
            model.add(Activation(activation)) # activation = iterating
            model.add(Dropout(0.3)) # Regularization parameter
            
        else:
            model.add(Dense(nodes)) # nodes = iterating
            model.add(Activation(activation)) # activation = iterating
            model.add(Dropout(0.3)) # Regularization parameter
    
    # output layer 
    model.add(Dense(units=1, 
                    kernel_initializer="glorot_uniform",
                   activation="sigmoid"))
    
        # compiling the model
    model.compile(optimizer="adam",
                  loss="binary_crossentropy",# for binary calssification
                 metrics=["accuracy"])
    return model

model = KerasClassifier(build_fn=create_model, verbose=0) # we have to define the model
```

```python
layers = [[20],[40,20],[45,30,15]] # number of layers
activations = ["sigmoid","relu"] # activation functions 
# sequence for 20 it will use sigmoid then relu

param_grid = dict(layers=layers, 
              activation=activations, 
              batch_size=[128,256], 
              epochs=[30])

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)
#cross validation of 5 folds
```

```python
grid_result = grid.fit(X_train, y_train)
```

```python
[grid_result.best_score_,grid_result.best_params_]
```

```python
# making predictions
y_pred = grid.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```
